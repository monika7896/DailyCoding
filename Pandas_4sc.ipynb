```python
import numpy as np
import pandas as pd
```

```python
data.drop(columns = ['director_id', 'id_y'], inplace = True)
```

```python
#risky director
#if a movie has higher budget than avg for that director => risky
```

```python
def risky(x):
    x['risky'] =  x['budget']  -  x['budget'].mean() > 0
    return x
```

```python
data.groupby('director_name',group_keys=True).apply(risky)
```

```python
data.groupby('director_name',group_keys=False).apply(risky)
```

```python
data_agg.columns = ['year_min','year_max','title_count']
```

```python
data_agg
```

```python
data_agg = data_agg.reset_index()
data_agg
```

```python
data_agg['active_year'] = data_agg['year_max'] - data_agg['year_min']
```

```python
data_agg
```

```python
data_agg['movies_per_year'] = data_agg['title_count']/data_agg['active_year']
```

```python
data_agg
```

```python
data_agg.sort_values(by='movies_per_year',ascending=False)
```
